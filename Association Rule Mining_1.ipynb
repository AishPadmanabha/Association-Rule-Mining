## 1. Online Retail Market Based Analysis

### **1.1 Frequent Itemset Generation**

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
pd.set_option("max_columns", 150)
pd.set_option("max_rows", 150)
```

```python
input_file = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
df = pd.read_excel(input_file)
basket = (df[df["Country"] == "United Kingdom"]
.groupby(["InvoiceNo", "Description"])["Quantity"]
.sum().unstack().reset_index().fillna(0)
.set_index("InvoiceNo"))
```

```python
basket_sets = basket.applymap(lambda x: True if x >=1 else False)
```

```python
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

#### frequent_itemsetsTop 5  1-itemsets with highest support

```python
frequent_itemsets_1 = frequent_itemsets.loc[frequent_itemsets['length'] == 1]
frequent_itemsets_1.sort_values(by='support',ascending=False)[:5].reset_index()
```

#### Top 5  2-itemsets with highest support

```python
frequent_itemsets_2 = frequent_itemsets.loc[frequent_itemsets['length'] == 2]
frequent_itemsets_2.sort_values(by='support',ascending=False)[:5].reset_index()
```

#### Highest support value for 1-itemset

```python
frequent_itemsets_1.sort_values(by='support',ascending=False)['support'].iloc[0]
```

#### Highest support value for 2-itemset

```python
frequent_itemsets_2.sort_values(by='support',ascending=False)['support'].iloc[0]
```

### **1.2 Association Rule Generation**

#### **Top 5 Association Rules**

Below the rules representation: one of the top association rules, and whether or not it makes sense:

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules.sort_values(by='confidence', ascending=False).reset_index()[:5]
```

The top rule is that one who gets pink regency cup and saucer is likely to get green regency cup and saucer. It makes sense if the person is wanting variety in their dining equipment. I can't think of any other reason that implication would be listed.
